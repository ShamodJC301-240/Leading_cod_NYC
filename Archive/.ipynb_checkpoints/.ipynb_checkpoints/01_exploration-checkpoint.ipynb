```python
# Setting root directory
BASE_DIR = Path.cwd().parent
# print(BASE_DIR)
RAW_DATA_DIR = BASE_DIR / "Data" / "raw"
# print(RAW_DATA_DIR)
raw_csv = RAW_DATA_DIR / "NY_leading_cause_of_death.csv"
# print(raw_csv)
```

```python
# Loading our data
Cause_Of_Death = pd.read_csv(raw_csv)

# Printing head
Cause_Of_Death.head()
```

### Note

In the dataset, the data types for `deaths`, `death_rate`, `age_adjusted_death_rate` are all labeled str when they are numerical values. Will change that soon to prevent issues in the future.

```python
# Adjusting data types
Cause_Of_Death["deaths"] = pd.to_numeric(Cause_Of_Death["deaths"], errors="coerce")

Cause_Of_Death["death_rate"] = pd.to_numeric(Cause_Of_Death["death_rate"], errors="coerce")

Cause_Of_Death["age_adjusted_death_rate"] = pd.to_numeric(Cause_Of_Death["age_adjusted_death_rate"], errors="coerce")

Cause_Of_Death.dtypes
```

```python
# Statistic summary of dataset
Cause_Of_Death.describe(include="all")
```

```python
# Checking for null values
Cause_Of_Death.isnull().sum().sort_values(ascending=False)
```

```python
# Pattern inspection for null values
Cause_Of_Death.isnull().value_counts()
```

## Interpretation

The missing values are not randomly distributed across the dataset and instead follow a clear pattern. A portion of records only lacks rate-based metrics (`death_rate` and `age_adjusted_death_rate`), while a smaller subset is missing all mortality-related fields, including `deaths`.

This suggests the presence of different types of missing values. Some rows likely reflect cases where rates could not be computed due to data limitations, while others appear to be structurally incomplete observations where no mortality metrics were recorded at all.

Additionally, the fact that many of these missing values were only revealed after converting columns to numeric types indicates that some invalid or non-standard entries were present in the original dataset and were not initially recognized as missing values.

Overall, the missing data appears to be systematic rather than random, which is important to consider when interpreting trends or performing further analysis.


```python
#Checking invalid death data
Cause_Of_Death[Cause_Of_Death["deaths"].isnull()]
```

## Missing Value Investigation

Inspection of rows with missing mortality metrics revealed a clear pattern: missing values were concentrated among smaller demographic groups such as `"Other Race/ Ethnicity"` and `"Not Stated/Unknown"`.

Because entire mortality-related fields (`deaths`, `death_rate`, and `age_adjusted_death_rate`) are simultaneously missing, this likely represents structural missingness rather than random data-entry error.

### Duplicates Interpretation

The dataset contains 10 duplicate rows. Further inspection is required to determine whether these duplicates are true data-entry redundancies or legitimate repeated records.

```python
# Present both original and duplicated rows
Cause_Of_Death[Cause_Of_Death.duplicated(keep=False)]
```

## Duplicate Record Inspection

The dataset contains 10 duplicate rows. Inspection of these records shows that duplicates primarily occur within the `"Other Race/ Ethnicity"` category and are associated with missing values in both the `death_rate` and `age_adjusted_death_rate` columns.

Many duplicated records involve very small death counts (e.g., 1–7 deaths), suggesting these duplicates may stem from data aggregation, reporting artifacts, or suppressed demographic calculations rather than random duplication.

```python
# Leading cause of death Frequency
Cause_Of_Death["leading_cause"].value_counts().head(10).plot(kind="bar")
plt.xticks(rotation=75)
plt.show()
```

```python
# Sex distribution
Cause_Of_Death["sex"].value_counts()
```

```python
# Race distribution
Cause_Of_Death["race_ethnicity"].value_counts()
```

```python
# Cleaning inconsistent categories
Cause_Of_Death["race_ethnicity"] = Cause_Of_Death["race_ethnicity"].replace({
    "White Non-Hispanic": "Non-Hispanic White",
    "Black Non-Hispanic": "Non-Hispanic Black"
})
```

```python
# Statistical summaries for numerical data
Cause_Of_Death[["deaths", "death_rate", "age_adjusted_death_rate"]].describe()
```

```python
# What are the highest death counts
Average_death_counts = (Cause_Of_Death.groupby("leading_cause")["deaths"].mean().sort_values(ascending=False).round(2))

print("Here are our average death counts")
print(Average_death_counts)
```

```python
# Highest_Death_Counts
Highest_Death_Counts = Cause_Of_Death.groupby("leading_cause")["deaths"].sum().sort_values(ascending=False).round(2)

print("Here are are highest death counts")
print(Highest_Death_Counts)
```
